### 문제 5
기존 `scores` 딕셔너리를 사용하여, 점수가 80점 이상이면 'Pass', 미만이면 'Fail'을 값으로 가지는 새로운 딕셔너리를 만드세요.

```python
# 코드를 여기에 작성하세요
students = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95,
    "Eve": 88
}

new_dict = {}

for name, score in students.items():
    if score >= 80:
        new_dict[name] = "Pass"
    else:
        new_dict[name] = "Fail"
    
print(new_dict)

# print(students)

for name in list(students.keys()):
    # print(students[name])
    score = students.pop(name)
    print(score)
    
    if score >= 80:
        students[name] = "Pass"
    else:
        students[name] = "Fail"
        
print(students)


    
```
