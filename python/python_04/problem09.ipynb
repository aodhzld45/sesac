### 문제 1
여러 명의 학생 성적 데이터가 담긴 복합 리스트입니다. 각 학생의 `grades` 리스트에서 80점 이상의 점수만 추출하여 평균을 계산하고, 그 평균이 90점 이상인 학생의 이름만 `top_students` 리스트에 담으세요.

```python
students = [
    {"name": "John", "grades": [85, 92, 78, 100]},
    {"name": "Sarah", "grades": [88, 95, 90, 89]},
    {"name": "Tyler", "grades": [70, 75, 82, 80]},
    {"name": "Emma", "grades": [95, 98, 88, 92]}
]

top_students = []
avg = 0.0
total = 0
for student in students:
    
    name = student["name"]
    print(name)
    
    total = sum(student["grades"])
    count = len(student["grades"])
    
    # print(f"총합 = {total} , 카운트 = {count}")
    
    avg = total / count
    
    print(f"평균 = {avg}")
    
    if avg >= 90:
        top_students.append(name)
        
print(top_students)
    
    
  


    
```
