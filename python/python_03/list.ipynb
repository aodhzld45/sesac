```python
# 반복문을 사용하여, 모든 학생들의 성적에 +30점의 가산점을 부여하라.
scores = [50, 70, 20, 40, 30, 60]

for num in range(len(scores)):
    score = scores[num]
    print("가산된", num + 1, "번째 학생의 성적 = ", score + 30)
    
```
