
### 문제 1
사용자로부터 정수 `n`을 입력받아, 1부터 `n`까지 모든 정수의 합을 출력하세요.

````python
n = int(input('정수를 입력하세요'))
i = 1
total = 0

while i <= n:
    print(i)
    total += i
    i += 1

print(total)
    
    
    
````

### 문제 4
`names = ['jun', 'ken', 'alex', 'sophia', 'neo']` 일 때,

출석번호를 포함하여 '1번 jun', ..., '5번 neo'의 형태로 출석을 부르시오.

````python

names = ['jun', 'ken', 'alex', 'sophia', 'neo']
i = 0
while i < len(names):
    print("출석번호:", i+1, "이름 : ", names[i])
    i += 1
    
````

### 문제 6

사용자로부터 숫자를 계속 입력받다가, `0`을 입력하면 지금까지 입력된 모든 숫자의 합계를 출력하고 종료하는 프로그램을 작성하세요.

````python
total = 0

while True: 
    n = int(input('정수를 입력하세요'))
    if n == 0:
        print("0이 입력되었습니다. 반복을 중지합니다.")
        break
    print("지금까지 입력된 정수 =", n)
    total += n
    
print("지금까지의 합계 =", total)

````
