### 문제 9
1부터 50까지의 숫자 중에서 5의 배수만 골라 `my_list`에 저장하고, 그 리스트의 모든 요소의 합을 `for`문을 사용하여 구하세요.

```python
my_list = []
total = 0

# 1. range를 사용하여 5의 배수를 리스트에 추가
# 2. for문을 사용하여 합계 계산

print(f"리스트: {my_list}")
print(f"합계: {total}")

for i in range(1, 50+1):
    if i % 5 == 0:
        print(i)
        my_list.append(i)
        
print(my_list)

for r in range(len(my_list)):
    print(r)
    total += my_list[r]
print(total)

    
```

### 문제 12
`while`문을 사용하여 1부터 100까지의 정수 중 홀수의 합과 짝수의 합을 각각 따로 구하여 출력하세요.

```python
odd_sum = 0
even_sum = 0
n = 1

while n <= 100:
    # 조건문을 사용하여 홀수/짝수 합계를 구하세요
    if n % 2 == 0:
        even_sum += n
    else:
        odd_sum += n
    n += 1
    
print(f"홀수의 합: {odd_sum}")
print(f"짝수의 합: {even_sum}")
```

### 문제 16

nums = [7, 2, 9, 9, 1, 5, 8, 8, 3] 일 때,
리스트에서 중복 없이 등장한 숫자만 뽑아서 unique 리스트로 만드세요.
(예: [7, 2, 9, 1, 5, 8, 3] 같은 형태)
(단, set() 사용 금지)

```python
nums = [7, 2, 9, 9, 1, 5, 8, 8, 3]
new_list = []
for num in nums:
    print(num)
    if num not in new_list:
        new_list.append(num)
        print(new_list)

print(new_list)
        
```

### 문제 17

for문을 사용해서 **1 ~ 100 사이의 소수(prime)**만 골라 리스트 primes에 담아 출력하세요.
(힌트: 어떤 수 x가 소수인지 판별하려면 2부터 x-1까지 나눠보는 방식)

```python
primes = []

for n in range(1, 100+1):
    prime = True
    
    for i in range(2, n):
        if n % i == 0:
            prime = False   
            break
        #else:    
            # print(prime, n)
    if prime == True:
        print(prime, n)        
        primes.append(n)
print(primes)

        
```

### 문제 18

while문을 사용해서 1부터 시작해 숫자를 계속 더할 때,
**누적합이 처음으로 500을 넘는 순간의 숫자(n)**와 그때의 **누적합(total)**을 출력하세요.
예: 1+2+3+...+n > 500을 처음 만족하는 n 찾기

```python
n = 1
total =0

while True:
    total += n
    
    if total > 500:
        print(n)
        print(total)
        break
    n += 1
    
```
