## 점심식사 프로그램 만들기

<pre>
    <b>조건들을 나열해보자</b>
    1. 최대 4명씩 조가 된다고 가정한다
    2. 총 인원이 4로 나누어 지지 않을 경우, 즉 초과나 미만일 경우 마지막 조에 편성하여 계산한다. 조건값 최소 3명 최대 4명 
    3. 만약 4명씩 조가 되었는데 해당 명단에서 오늘 결석자가 존재할 경우 리스트에서 제외한다.
    4. 어제 같이 식사한 교육생들은 오늘 식사에서 제외한다. -> 이 부분 같은 경우는 
    5. 같이 식사한 교육생들의 초기화 기간은 일주일로 산정한다.
</pre>


```python
import random

names = [
    "손오공", "베지터", "손오반", "피콜로", "크리링",
    "야무치", "천진반", "차오즈", "부르마", "트랭크스",
    "손오천", "치치", "비델", "미스터 사탄", "마인 부우",
    "프리저", "셀", "기뉴", "자봉", "도도리아",
]

def remove_absent_person(names):
    absent_people = []
    while True:
        absent_person = input('결석자 입력 \n 없으면 빈값(엔터)')

        if absent_person not in names:
            break
        if absent_person in names:
            names.remove(absent_person)
            absent_people.append(absent_person)

    print(f'결석자 : {absent_people}')

    return absent_people


def lunch(names, absent_list, size):
    absents = set(absent_list or [])
    clean_list = [n for n in names if n not in absents]

    random.shuffle(clean_list)
    
    groups = []
    current = []

    for n in clean_list:
        current.append(n)
        if len(current) == size:
            groups.append(current)
            current = []

    if current:
        groups.append(current)

    # 마지막 조가 1~2명이면 3명 이상 되도록 보정 (모든 조 3~4 유지)
    if len(groups) >= 2 and len(groups[-1]) < 3:
        last = groups[-1]

        if len(last) == 2:
            # [4,2] -> [3,3]
            last.append(groups[-2].pop())

        elif len(last) == 1:
            # [4,4,1] -> [3,3,3] (가능할 때)
            if len(groups) >= 3:
                last.append(groups[-2].pop())
                last.append(groups[-3].pop())
            else:
                print("[불가] 조 보정이 불가능한 케이스입니다. (인원 구성 확인 필요)")
                return None

    return groups
            
result = lunch(names, remove_absent_person(names), 4)

if result is not None:
    print("\n=== 오늘 점심 조 편성 ===")
    for idx, g in enumerate(result, start=1):
        print(f"{idx}조 = {g}")
        
    
```
